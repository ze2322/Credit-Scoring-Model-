```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict, KFold ,cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.dummy import DummyClassifier
%matplotlib inline
```

we need to repalce the - with 0

```python
data = data.replace({'-':0})
```

we will split  the data set into x and y as dependent and independent values

```python
y = data['Score_class']
X = data.drop(['Score_class'], axis=1, inplace=False)
```

view x

```python
X
```

```python
y
```

we will split the dataset into training and testing


```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.30)
```

view the shape of x train and x test

```python
print(X_train.shape, X_test.shape)
```

view the shape of y train and y test

```python
print(y_train.shape, y_test.shape)
```




*   we will use LogisticRegression as a binary classification to predict the output
*   we will also use cross validation to assess the preformance of the model
*   we will use radom state = 150


```python
num_folds = 5
model = LogisticRegression()
kf = KFold(n_splits=num_folds, shuffle=True, random_state=150)
cross_val_predictions = cross_val_predict(model, X_train, y_train, cv=kf)
```



*   we will calulate the accuracy
*   we will calulate the confusion matrix
*   we will calulate the classification report



```python
accuracy_cv = accuracy_score(y_train, cross_val_predictions)
conf_matrix_cv = confusion_matrix(y_train, cross_val_predictions)
class_report_cv = classification_report(y_train, cross_val_predictions)
```


*   we will print the accuracy
*   we will print the confusion matrix
*   we will print the classification report

```python
print('accuracy')
print(accuracy_cv)

print('conf matrix')
print(conf_matrix_cv)

print('class report')
print(class_report_cv)
```

we will make a dummy model to see cross validation score

```python
dummy_classifier = DummyClassifier()
dummy_classifier.fit(X_train, y_train)
dummy_predictions = dummy_classifier.predict(X_test)
```

```python
cv_scores = cross_val_score(dummy_classifier, X_train, y_train, cv=3)
print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())
```

now we will compare between actual data and predicting data

```python
result_df = pd.DataFrame({'Actual': y_train, 'Predicted': cross_val_predictions})

# Display the DataFrame for the entire dataset
print("Results on Entire Dataset:")
print(result_df.head(50))
```
